### Import libraries

```python
import numpy as np
import pandas as pd

#Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#Train test splitting
from sklearn.model_selection import train_test_split

# Model
from sklearn.linear_model import LogisticRegression

#Evaluation
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df.info()
```

```python
df['Species'].unique()
```

```python
df['Species'] = df['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1,
                                      'Iris-virginica':2})
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['Id'], axis=1, inplace=True)
```

```python
df.info()
```

### Split the data

```python
X = df.drop(['Species'], axis=1)
y = df['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    random_state=1)
```

### Model Building

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
y_pred
```

```python
# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
#confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import multilabel_confusion_matrix
```

```python
multilabel_confusion_matrix(y_test, y_pred)
```

```python
#classification
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
logi_reg1 = LogisticRegression(multi_class='ovr')  #One vs Rest
logi_reg1.fit(X_train, y_train)
```

```python
y_pred_ovr = logi_reg1.predict(X_test)
y_pred_ovr
```

```python
# Accuracy Score
accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
accuracy_ovr
```

```python
multilabel_confusion_matrix(y_test, y_pred_ovr)
```

```python
 #classification
class_report = classification_report(y_test, y_pred_ovr)
print(class_report)
```
